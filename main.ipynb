```python
dataframe_movies = dataframe_movies[[ 'id',  'title', 'genres', 'cast', 'crew','overview', 'keywords']]
```

```python
import ast

def convert(data):
    genre  = []

    for i in ast.literal_eval(data):
        # temp = i.name
        genre.append(i['name'])
    
    return genre

dataframe_movies['genres'] = dataframe_movies['genres'].apply(convert)
dataframe_movies.head(5)
```

```python

dataframe_movies['keywords'] = dataframe_movies['keywords'].apply(convert)
```

```python
def convert_cast(data):
    l = []
    counter = 0 
    for i in ast.literal_eval(data):
        if counter != 3 :
            l.append(i['name'])
            counter += 1
        else:
            break
    return l 

dataframe_movies['cast'] = dataframe_movies['cast'].apply(convert_cast)
```

```python
dataframe_movies.head(1)
```

```python
dataframe_movies['crew'].iloc[0]
```

```python
def fetch_director(data):
    l = []
    for i in ast.literal_eval(data):
        if i['job'] == 'Director':
            l.append(i['name'])
    return l 

dataframe_movies['crew'] = dataframe_movies['crew'].apply(fetch_director)
```

```python
dataframe_movies.head(1)
```

```python
dataframe_movies['overview'] = dataframe_movies['overview'].apply(lambda x : x.split())
dataframe_movies.head(1)
```

```python
# we'll concatenate the strings becase james cameron will be interpreted as different entity not a single entity for this we will apply transformation 

dataframe_movies['genres'] = dataframe_movies['genres'].apply(lambda x : [i.replace(" ", '') for i in x])
dataframe_movies['keywords'] = dataframe_movies['keywords'].apply(lambda x : [i.replace(" ", '') for i in x])
dataframe_movies['cast'] = dataframe_movies['cast'].apply(lambda x : [i.replace(" ", '') for i in x])
dataframe_movies['crew'] = dataframe_movies['crew'].apply(lambda x : [i.replace(" ", '') for i in x])
```

```python
dataframe_movies['tags'] = dataframe_movies['genres'] + dataframe_movies['keywords'] + dataframe_movies['cast'] + dataframe_movies['crew']
```

```python
dataframe_movies.head(1)
```

```python
dataframe = dataframe_movies[['id', 'title','tags']]
```

```python
dataframe.head(1)
```

```python
dataframe['tags'] = dataframe['tags'].apply(lambda x : " ".join(x))
```

```python
dataframe.iloc[0]['tags']
```

```python
dataframe['tags'] = dataframe['tags'].apply(lambda x : x.lower())
```

```python
# now we'll perform the text vectorization . 
```

```python
from nltk.stem import PorterStemmer

ps = PorterStemmer()

def stem_data(text):
    lst = []
    for i in text.split() :
        lst.append(ps.stem(i))

    return " ".join(lst)

dataframe['tags'] = dataframe['tags'].apply(stem_data)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
dataframe.head(1)
```

```python
vector =  cv.fit_transform(dataframe['tags']).toarray()
```

```python
dataframe
```

```python
len(cv.get_feature_names_out())
```

```python
# now we find the distance here we will use cosine distance . at higher dimension euclidian distance is not a reliable solution so we ll use cosine distance
```

```python
from sklearn.metrics.pairwise import cosine_similarity 

similarity = cosine_similarity(vector)
```

```python
sorted(list(enumerate(similarity[0])), reverse=True, key = lambda x : x[1] ) # always hold the index simlirarity[0] mean . similiarity with respect to movie 0 
```

```python
# fetch the index of the movie 

dataframe[dataframe['title'] == 'Batman Begins'].index[0]
```

```python
def recommend(movie):
    movie_list = []
    
    movie_index = dataframe[dataframe['title'] == movie].index[0]
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)), reverse=True, key = lambda x : x[1] )[1:6]

    for i in movie_list :
        print(dataframe.iloc[i[0]]['title'])
    return

recommend('Batman')
```
